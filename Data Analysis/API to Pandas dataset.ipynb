```python
import requests
import numpy as np
import pandas as pd
```

```python
response = requests.get("https://api.themoviedb.org/3/movie/popular?api_key=e83824d26002f05a9d41e50e01e2c1c6&language=en-US&page=1").json()
```

```python
response
```

### This is the for the first page of the API call

```python
popularity = []
vote_count = []
original_lang = []
title = []
vote_avg = []
overview = []
release_date = []

for i in response['results']:
    popularity.append(i['popularity'])
    vote_count.append(i['vote_count'])
    original_lang.append(i['original_language'])
    title.append(i['title'])
    vote_avg.append(i['overview'])
    release_date.append(i['release_date'])
    
d= {'title':title, 'overview':overview,
    'original_lang':original_lang,
    'release_date':release_date, 'popularity':popularity,
    'vote_count':vote_count, 'vote_avg':vote_avg}

df = pd.DataFrame.from_dict(d, orient= 'index')
df = df.transpose()
```

```python
final = pd.DataFrame()

for j in range(1, 30):
    
    response = requests.get("https://api.themoviedb.org/3/movie/popular?api_key=e83824d26002f05a9d41e50e01e2c1c6&language=en-US&page={}".format(j)).json()
    
    popularity = []
    vote_count = []
    original_lang = []
    title = []
    vote_avg = []
    overview = []
    release_date = []

    for i in response['results']:
        popularity.append(i['popularity'])
        vote_count.append(i['vote_count'])
        original_lang.append(i['original_language'])
        title.append(i['title'])
        vote_avg.append(i['overview'])
        release_date.append(i['release_date'])

    d= {'title':title, 'overview':overview,
        'original_lang':original_lang,
        'release_date':release_date, 'popularity':popularity,
        'vote_count':vote_count, 'vote_avg':vote_avg}

    df = pd.DataFrame.from_dict(d, orient= 'index')
    df = df.transpose()
    
    final = final.append(df, ignore_index = True)
```
