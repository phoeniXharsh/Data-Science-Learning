```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df= pd.read_json('https://www.mohfw.gov.in/data/datanew.json')
```

```python
df
```

```python
df = df.iloc[:-1, 1:]
```

```python
df
```

```python
total_states = np.arange(len(df['state_name']))
```

```python
total_states
```

### Total no.of positive cases for all states

```python
from matplotlib.pyplot import figure
figure(num= None, figsize= (12,8), dpi=80,facecolor= 'w', edgecolor= 'k')
plt.barh(total_states, df['positive'], align= 'center', alpha= 0.5,
                color=(1,0,0), edgecolor= (0.5, 0.2, 0.8))
plt.yticks(total_states, df['state_name'])
# plt.xlim(1, max(df['positive']))
plt.xlabel('Positive Number of Cases')
plt.title('Corona Virus Cases')
plt.show()
```

### Active number of cases based on states

```python
from matplotlib.pyplot import figure
figure(num= None, figsize= (12,8), dpi=80,facecolor= 'w', edgecolor= 'k')
plt.barh(total_states, df['positive'], align= 'center', alpha= 0.5,
                color=(1,0.5,0), edgecolor= (0.5, 0.2, 0.8))
plt.yticks(total_states, df['state_name'])
# plt.xlim(1, max(df['active'])+10000000)
plt.xlabel('Active Number of Cases')
plt.title('Corona Virus Cases')
plt.show()
```

### Death number of cases

```python
df.groupby('state_name')['death'].sum()
```

```python
from matplotlib.pyplot import figure
figure(num= None, figsize= (12,8), dpi=80,facecolor= 'w', edgecolor= 'k')
plt.barh(total_states, df['positive'], align= 'center', alpha= 0.5,
                color=(0,0,1), edgecolor= (0.5, 0.2, 0.8))
plt.yticks(total_states, df['state_name'])
# plt.xlim(1, max(df['death'])+8000000)
plt.xlabel('Death Number of Cases')
plt.title('Corona Virus Cases')
plt.show()
```
