## Count or Frequency Encoding

```python
train_set = pd.read_csv('https://archive.ics.uci.edu/ml/machine-learning-databases/adult/adult.data', header = None,index_col=None)
train_set.head()
```

```python
columns=[1,3,5,6,7,8,9,13]
```

```python
train_set= train_set[columns]
```

```python
train_set.columns=['Employment', 'Degree', 'Status', 'Designation', 'Family_job', 'Race', 'Sex', 'Country']
```

```python
train_set.head()
```

```python
for feature in train_set.columns:
    print(feature, ":", len(train_set[feature].unique()), 'labels')
```

```python
train_set['Country'].value_counts().to_dict()
```

```python
country_map = train_set['Country'].value_counts().to_dict()
```

```python
train_set['Country']= train_set['Country'].map(country_map)
train_set.head(20)
```

#### Advantages 
1. Easy to Use
2. Not increasing feature space

#### Disadvantages 
1. It will provide the same weight if the frequencies are same

## Target Guided Ordinal Encoding
1. Ordering the labels according to the target
2. Raplce the labels by the joint probability of being 1 or 0

```python
df= pd.read_csv('Datasets/titanic_train.csv', usecols= ['Cabin', 'Survived'])
df.head()
```

```python
df['Cabin'].fillna('Missing', inplace= True)
```

```python
df.head()
```

```python
df['Cabin']= df['Cabin'].astype(str).str[0]
df.head()
```

```python
df.groupby(['Cabin'])['Survived'].mean().sort_values()
```

```python
df.groupby(['Cabin'])['Survived'].mean().sort_values().index
```

```python
ordinal_labels= df.groupby(['Cabin'])['Survived'].mean().sort_values().index
ordinal_labels
```

```python
# Assinging a number to all the lablels
ordianl_labels2= {k:i for i,k in enumerate(ordinal_labels, 0)}
ordianl_labels2
```

```python
df['Cabin_ordinal_labels']= df['Cabin'].map(ordianl_labels2)
df.head()
```

## Mean Encoding

```python
df.groupby(['Cabin'])['Survived'].mean().to_dict()
```

```python
mean_ordinal= df.groupby(['Cabin'])['Survived'].mean().sort_values()
```

```python
df['mean_ordinal_encode']= df['Cabin'].map(mean_ordinal)
df.head()
```

#### Advantages
1. Captures the info within the label therefore rendering more predictive features
2. Creates monotonic relationship between variable and the target

#### Disadvantages
1. Prone to overfitting

## Probability Ratio Encoding 
1. Probability of survived based on Cabin : categorical feature
2. Probability of Not survived : 1 - pr(survived)
3. pr(survived)/pr(not_survived)
4. dictionary to map cabin with probability of survival
5. replace the categorical feature with the probability

```python
df= pd.read_csv('Datasets/titanic_train.csv', usecols= ['Cabin', 'Survived'])
df.head()
```

```python
## Raplacing 
df['Cabin'].fillna('Missing', inplace = True)
df.head()
```

```python
df['Cabin'].unique()
```

```python
df['Cabin']= df['Cabin'].astype(str).str[0]
df.head()
```

```python
df.Cabin.unique()
```

```python
prob_df= df.groupby(['Cabin'])['Survived'].mean()
```

```python
prob_df = pd.DataFrame(prob_df)
prob_df.head()
```

```python
prob_df['Died']= 1 - prob_df['Survived']
```

```python
prob_df.head()
```

```python
prob_df['Probability_ratio'] = prob_df['Survived']/prob_df['Died']
```

```python
prob_df.head()
```

```python
prob_encoded = prob_df['Probability_ratio'].to_dict()
```

```python
df['Cabin_encoded']= df['Cabin'].map(prob_encoded)
df.head(20)
```
